```python
# =========================================================
# ONE-CELL PIPELINE (NO UNZIP)
# 1) Read already-extracted folders
# 2) Build dataframe from Parasitized + unaffected/uninfected
# 3) Split by group_id (avoid leakage)
# 4) Train EfficientNetB0
# 5) Evaluate
# =========================================================

import os
import re
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# -----------------------------
# SETTINGS
# -----------------------------
data_root = "/kaggle/input/datasets/hajarakandeh/malaria-augmented-updated-data"   # change if needed
img_size = (224, 224)
batch_size = 16
epochs = 20
seed = 42

AUTOTUNE = tf.data.AUTOTUNE
np.random.seed(seed)
tf.random.set_seed(seed)

# -----------------------------
# FIND CLASS FOLDERS
# -----------------------------
def find_folder(base_dir, target_name):
    for root, dirs, files in os.walk(base_dir):
        if os.path.basename(root).lower() == target_name.lower():
            return root
    return None

parasitized_dir = find_folder(data_root, "Parasitized")

# try both names because you mentioned "unaffected" before,
# but earlier code also used "uninfected"
negative_dir = find_folder(data_root, "unaffected")
if negative_dir is None:
    negative_dir = find_folder(data_root, "uninfected")

print("Parasitized folder:", parasitized_dir)
print("Negative folder   :", negative_dir)

if parasitized_dir is None or negative_dir is None:
    raise FileNotFoundError(
        "Could not find 'Parasitized' and 'unaffected'/'uninfected' folders."
    )

# -----------------------------
# BUILD LABEL DATAFRAME
# -----------------------------
valid_ext = (".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff")
records = []

folder_label_map = {
    parasitized_dir: "infected",
    negative_dir: "negative"
}

for folder_path, label_name in folder_label_map.items():
    for fname in os.listdir(folder_path):
        if not fname.lower().endswith(valid_ext):
            continue

        fpath = os.path.join(folder_path, fname)
        stem = os.path.splitext(fname)[0]

        aug_match = re.search(r"_aug(\d+)$", stem)
        if aug_match:
            img_type = f"aug{aug_match.group(1)}"
            group_id = re.sub(r"_aug\d+$", "", stem)
        else:
            img_type = "original"
            group_id = stem

        records.append({
            "filepath": fpath,
            "label": label_name,
            "type": img_type,
            "group_id": group_id
        })

df = pd.DataFrame(records)

if df.empty:
    raise ValueError("No images found in the dataset folders.")

df["label_num"] = df["label"].map({"negative": 0, "infected": 1})
df["exists"] = df["filepath"].apply(os.path.exists)

print("\nPath check:")
print(df["exists"].value_counts())

if (~df["exists"]).any():
    raise ValueError("Some image paths do not exist.")

print("\nLabel counts:")
print(df["label"].value_counts())

print("\nType counts:")
print(df["type"].value_counts())

print("\nUnique groups:", df["group_id"].nunique())

# optional save of labels
label_csv_path = "/kaggle/working/labels_generated.csv"
df.to_csv(label_csv_path, index=False)
print("\nGenerated labels saved to:", label_csv_path)

# -----------------------------
# GROUPED SPLIT TO AVOID LEAKAGE
# -----------------------------
group_df = df[["group_id", "label_num"]].drop_duplicates()

train_groups, temp_groups = train_test_split(
    group_df,
    test_size=0.30,
    stratify=group_df["label_num"],
    random_state=seed
)

val_groups, test_groups = train_test_split(
    temp_groups,
    test_size=0.50,
    stratify=temp_groups["label_num"],
    random_state=seed
)

train_df = df[df["group_id"].isin(train_groups["group_id"])].copy()
val_df   = df[df["group_id"].isin(val_groups["group_id"])].copy()
test_df  = df[df["group_id"].isin(test_groups["group_id"])].copy()

print("\nSplit sizes:")
print("Train:", len(train_df))
print("Val  :", len(val_df))
print("Test :", len(test_df))

# -----------------------------
# TF DATASET
# -----------------------------
def load_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, img_size)
    img = tf.cast(img, tf.float32)
    img = tf.keras.applications.efficientnet.preprocess_input(img)
    return img, label

def make_dataset(dataframe, training=False):
    paths = dataframe["filepath"].values
    labels = dataframe["label_num"].values.astype("float32")

    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(load_image, num_parallel_calls=AUTOTUNE)

    if training:
        ds = ds.shuffle(min(len(dataframe), 10000), seed=seed)

    ds = ds.batch(batch_size).prefetch(AUTOTUNE)
    return ds

train_ds = make_dataset(train_df, training=True)
val_ds   = make_dataset(val_df, training=False)
test_ds  = make_dataset(test_df, training=False)

# -----------------------------
# MODEL: EFFICIENTNETB0
# -----------------------------
base_model = tf.keras.applications.EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(img_size[0], img_size[1], 3)
)
base_model.trainable = False

inputs = tf.keras.Input(shape=(img_size[0], img_size[1], 3))
x = base_model(inputs, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)

model = tf.keras.Model(inputs, outputs)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="binary_crossentropy",
    metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]
)

model.summary()

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=3,
        restore_best_weights=True
    )
]

# -----------------------------
# TRAIN
# -----------------------------
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs,
    callbacks=callbacks
)

# -----------------------------
# OPTIONAL FINE-TUNING
# -----------------------------
base_model.trainable = True

for layer in base_model.layers[:-20]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss="binary_crossentropy",
    metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]
)

history_ft = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=5,
    callbacks=callbacks
)

# -----------------------------
# EVALUATION
# -----------------------------
test_loss, test_acc, test_auc = model.evaluate(test_ds, verbose=1)
print(f"\nTest Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
print(f"Test AUC: {test_auc:.4f}")

y_true = test_df["label_num"].values
y_prob = model.predict(test_ds).ravel()
y_pred = (y_prob >= 0.5).astype(int)

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=["negative", "infected"]))

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["negative", "infected"])
disp.plot()
plt.title("Confusion Matrix")
plt.show()

# -----------------------------
# PLOT TRAINING HISTORY
# -----------------------------
def plot_history(histories, metric, title):
    plt.figure(figsize=(7, 5))
    for h, label in histories:
        if metric in h.history:
            plt.plot(h.history[metric], label=f"{label} train")
        val_metric = f"val_{metric}"
        if val_metric in h.history:
            plt.plot(h.history[val_metric], label=f"{label} val")
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel(metric)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_history([(history, "phase1"), (history_ft, "phase2")], "accuracy", "Model Accuracy")
plot_history([(history, "phase1"), (history_ft, "phase2")], "loss", "Model Loss")
plot_history([(history, "phase1"), (history_ft, "phase2")], "auc", "Model AUC")

# -----------------------------
# SAVE MODEL
# -----------------------------
save_path = "/kaggle/working/efficientnet_malaria_model.keras"
model.save(save_path)
print("\nModel saved to:", save_path)
```
